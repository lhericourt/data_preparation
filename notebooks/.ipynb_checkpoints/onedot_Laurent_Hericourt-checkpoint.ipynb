```python
import pandas as pd
import numpy as np
import json
```

# 0 - Pre-processing

```python
# Import of supplier file
SUPPLIER_DATA_PATH = "/Users/lolo/Développement/Python/onedot/data/supplier_car.json"
with open(SUPPLIER_DATA_PATH) as f:
    supplier_data_str = f.readlines()
supplier_data_json = [json.loads(x.strip()) for x in supplier_data_str]
```

```python
# Load supplier data into a dataframe
supplier_data = pd.DataFrame(supplier_data_json)
supplier_data = supplier_data.sort_values(by="ID")
supplier_data.head()
```

```python
# Set the values of column "Attribute Names" as different columns : one column per value 
supplier_data_as_table = supplier_data.pivot(index= "ID", columns="Attribute Names", values="Attribute Values").reset_index()

# Add general information
supplier_data_as_table = pd.merge(supplier_data_as_table, supplier_data[["ID", "MakeText", "ModelText", "ModelTypeText", "TypeName", "TypeNameFull"]].drop_duplicates(subset="ID"), on="ID", how="left")
supplier_data_as_table.head()
```

```python
# Replace value "null" by the pandas null value : Nan
supplier_data_as_table.fillna(value=np.NaN, inplace=True)
supplier_data_as_table = supplier_data_as_table.replace("null", np.nan)
supplier_data_as_table.head()
```

## Rule to compute the car type
We can use serveral attributes to find out the car type as defined in the database.

Here the rule we can set :

- If the field "Properties" contains the string "Tuning" then the value of the carType is "Custom"

- Else if the field "Seats" equal to "1" then the value of the carType is "Single Seater"

- Else we use the type mapping dictionary

## Rule to compute the make
To compute the make of one car we check if it exists in the database. 
- If so we use the case used in the database
- If not we use directly the make provided by the supplier but we warn the user that this make is not an existing make of the database

## Rule to compute the model

We use the make, the model and the model variant of the database to compute the model of the cars from the json.

Indeed the models in the json are usually more precised than the models in the database. Also we need to find out at what model of the database corresponds the model of the json.

- For a model of the json : we split it by space, to get a list of unique word. Then we compute all the combination of this words
- For the model of the database we concatenate the model and the model variant, and we delete the spaces.

Then we check if one combination of the words is contained in the concatenation of the models and variants of the car of the same make :
- If so we return the model of this car.
- If not we use directly the model provided by the supplier but we warn the user that this model is not an existing model of the database


## Rule to compute the model variant

The model variant in the supplier data sometimes repeat the model at the beginning : also to avoid information duplication we remove from the variant the model when it is present at the beginning of the variant
